```python
def get_stander_dict(history):
    history_list = history.split("\n")[1:-1]

    history_list = [line.split(",") for line in history_list]

    stander_dict = {"Loss":[], "Accuracy":[], "Test Loss":[], "Test Accuracy":[]}
    for line_list in history_list:
        for item in line_list[1:]:
            item_name, item_value = item.split(":")[0].strip(), float(item.split(":")[1].strip())
            stander_dict[item_name].append(item_value)
    return stander_dict
```
